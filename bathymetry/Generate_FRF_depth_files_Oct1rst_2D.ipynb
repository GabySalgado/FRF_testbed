```python
%matplotlib inline
import scipy
import numpy as np
import os, re
from scipy.interpolate import griddata, interp2d, interp1d
import matplotlib.pyplot as plt
```

```python
ProjectDir = os.getcwd()  # get current path
FolderName = 'DEPTH_FILES' # folder name where DEPTH files are stored
DepthDir = os.path.join(ProjectDir,FolderName) # path to FolderName

# Upload Oct 1 2D bathy
elevOct1 = np.loadtxt(os.path.join(DepthDir,'Oct1_FRF','elev_oct1'),delimiter=',')
[Oct1Ny,Oct1Nx] = np.shape(elevOct1) # x = number of columns in Depth10m and y = number of rows in elevOct1 

XshoreOct1 = np.loadtxt(os.path.join(DepthDir,'Oct1_FRF','FRF_xshore_Oct1'),delimiter=',')
DxOct1 = 10.0
YshoreOct1 = np.loadtxt(os.path.join(DepthDir,'Oct1_FRF','FRF_yshore_Oct1'),delimiter=',')
DyOct1 = 10.0

# Upload Oct 1 extended profile
OldProfile = np.loadtxt(os.path.join(DepthDir,'depth_1D_2015-10-01.txt')) *-1
OldDx = 1.0 
xOld = np.arange(0,len(OldProfile[1,:])*OldDx,OldDx)
OldProfileIntersect = 500.0 #OldProfile intersect to elevOct1

# Upload LIDAr Foreshores by date  
foreshoreDir = os.path.join(DepthDir,'MEAN_FORESHORE') # import foreshore x & z vectors for each date
foreshoreList = []
XforeshoreList = []

for f in os.listdir((foreshoreDir)):
    if re.match('Foreshore', f):
        foreshoreList.append(f)
    elif re.match('Xforeshore', f):
        XforeshoreList.append(f)

foreshoreList = sorted(foreshoreList)
XforeshoreList = sorted(XforeshoreList)

# Upload foreshore - bathy intersect by date
ForeshoreIntersect = np.loadtxt(os.path.join(DepthDir,'foreshore_Oct1_intersec2D.txt'))  

# Final Bathy grid resolution 
newDx = 1.0 
newDy = 2.0 
```

```python
for i in range(len(foreshoreList)):  
    date = foreshoreList[i][10:21].replace("-", "") 
    print('loading date:',date)
    
    foreshore = np.loadtxt(os.path.join(foreshoreDir,foreshoreList[i]))
    Xforeshore = np.loadtxt(os.path.join(foreshoreDir,XforeshoreList[i]))
    
    
    # find foreshore - bathy intercept index
    ForeshoreIndex = np.argmin(abs(Xforeshore-ForeshoreIntersect[i])) # index of closest value to intercept in Xforeshore vector
    BathyIndex = np.argmin(abs(XshoreOct1-ForeshoreIntersect[i])) # index of closest value to intercept 
    
    xConcatenateForeshore2elevOct1 = np.concatenate((Xforeshore[:ForeshoreIndex], XshoreOct1[BathyIndex:]), axis=0) # concatenate vector sections
    
    DepthConcatenateForeshore2elevOct1 = np.zeros([Oct1Ny,len(xConcatenateForeshore2elevOct1)]) # concatenate forshore to Depth10m cross-shore profiles
    for j in range(Oct1Ny):
        DepthConcatenateForeshore2elevOct1[j,:] = np.concatenate((foreshore[:ForeshoreIndex], elevOct1[j,BathyIndex:]), axis=0)
    
    f_interp1 = interp2d(xConcatenateForeshore2elevOct1,YshoreOct1,DepthConcatenateForeshore2elevOct1, kind='linear')
    
    
    XinterpOriginal = np.arange(xConcatenateForeshore2elevOct1[0],xConcatenateForeshore2elevOct1[-1],newDx) # in original FRF coordinates, but with newDx
    YinterpOriginal = np.arange(YshoreOct1[0],YshoreOct1[-1],newDy)
    
    DepthConcatenate2Oct1stBathy = f_interp1(XinterpOriginal,YinterpOriginal)
    
    
    # change to FUNWAVE local coordinates:
    nonNanForshoreIndex = int(np.argwhere(np.isfinite(DepthConcatenate2Oct1stBathy[1,:30]))[0])
    
    
    NewDepthX = np.arange(0,len(XinterpOriginal[nonNanForshoreIndex:])*newDx,newDx)
    NewDepthY = np.arange(0,len(YinterpOriginal)*newDy,newDy)
    NewDepth = np.zeros([len(YinterpOriginal),len(XinterpOriginal[nonNanForshoreIndex:])])                     
    NewDepth[:,:] = DepthConcatenate2Oct1stBathy[:,nonNanForshoreIndex:]
    
    
    # find newDepth - Oct1stProfile intercept index
    ConcatenateIndex = np.argmin(abs(NewDepthX-OldProfileIntersect)) # index of closest value to intercept in Xforeshore vector
    OldIndex = np.argmin(abs(xOld-OldProfileIntersect)) # index of closest value to intercept in Depth10mX
    
    yConcatenate2Oct1stProfile = NewDepthY
    xConcatenate2Oct1stProfile = np.concatenate((NewDepthX[:ConcatenateIndex], xOld[OldIndex:]), axis=0) # concatenate X vectors
    DepthConcatenate2Oct1stProfile = np.zeros([len(NewDepthY),len(xConcatenate2Oct1stProfile)]) 
    
    # concatenate newDepth to Oct 1st Profile
    for k in range(len(NewDepthY)):
        DepthConcatenate2Oct1stProfile[k,:] = np.concatenate((NewDepth[k,:ConcatenateIndex], OldProfile[1,OldIndex:]), axis=0)
        
    (NyConcatenate2Oct1stProfile,NxConcatenate2Oct1stProfile) = DepthConcatenate2Oct1stProfile.shape    
    
    
    ## Save 1D profiles
    Profile = np.zeros([3,NxConcatenate2Oct1stProfile])
    for l in range(3):
        Profile[l,:] = DepthConcatenate2Oct1stProfile[36,:] * -1 
    
    fileOut1D = os.path.join(ProjectDir,'1D_profiles','depth1D_Oct1stSurvey_%smglob_%s.txt'%(str(len(xConcatenate2Oct1stProfile)),str(date)))
    print('Saving 1D File:',fileOut1D)
    np.savetxt(fileOut1D, Profile, fmt='%.6e', delimiter=' ', newline='\n', header='', footer='', comments='! ')
    plt.plot(Profile[1,:])
     
    ## extend offshore for 2D bathy
    OffshoreSecLength = 100.0
    extStart = DepthConcatenate2Oct1stProfile[:,-1]
    xOffshore = np.arange(0,OffshoreSecLength+newDx,newDx) 
    OffshoreProfileNum = int(OffshoreSecLength/newDx)
    
    OffshoreSec = np.zeros([NyConcatenate2Oct1stProfile,OffshoreProfileNum])
    for i in range(OffshoreProfileNum):
        OffshoreSec[:,i] = extStart

    # merge to bathy
    OffshoreDep = np.zeros([NyConcatenate2Oct1stProfile,NxConcatenate2Oct1stProfile+OffshoreProfileNum])
    OffshoreDep[:,:NxConcatenate2Oct1stProfile] = DepthConcatenate2Oct1stProfile
    OffshoreDep[:,NxConcatenate2Oct1stProfile:] = OffshoreSec
    
    
    (NyExtendedOffshore,NxExtendedOffshore) = OffshoreDep.shape
    LxExtendedOffshore = NxExtendedOffshore*newDx
    LyExtendedOffshore = NyExtendedOffshore*newDy
    xExtendedOffshore = np.linspace(0,LxExtendedOffshore,NxExtendedOffshore)
    yExtendedOffshore = np.linspace(0,LyExtendedOffshore,NyExtendedOffshore)

    ## extend periodic boundary at the north for 2D bathy
    PeriodicSecLength = 100.0
    ProfileStart = OffshoreDep[0,:]
    ProfileEnd = OffshoreDep[-1,:]
    
    # create periodic boundary sections 
    PeriodicSecY = np.array([0,PeriodicSecLength])
    PeriodicProfileNum = len(PeriodicSecY)

    PeriodicSec = np.zeros([PeriodicProfileNum,NxExtendedOffshore])
    PeriodicSec[0,:] = ProfileEnd
    PeriodicSec[1,:] = ProfileStart

    
    # interpolate 
    f_periodic = interp2d(xExtendedOffshore,PeriodicSecY,PeriodicSec, kind='linear')

    NewPeriodicSecY = np.arange(0,PeriodicSecLength+newDy,newDy)
    NewPeriodicSec = f_periodic(xExtendedOffshore,NewPeriodicSecY)
    
    # merge to FINAL bathy
    FinalDepth = np.zeros([NyExtendedOffshore+len(NewPeriodicSecY),NxExtendedOffshore])
    (FinalNy,FinalNx) = FinalDepth.shape
    FinalLx = FinalNx*newDx
    FinalLy = FinalNy*newDy
    xFinal = np.linspace(0,FinalLx,FinalNx)
    yFinal = np.linspace(0,FinalLy,FinalNy)
    
    FinalDepth[:NyExtendedOffshore,:] = OffshoreDep[:,:]*-1
    FinalDepth[NyExtendedOffshore:,:] = NewPeriodicSec*-1
    
    PeriodicProfile = np.zeros(FinalNx) + (NyExtendedOffshore*newDy)
    OffshoreProfile = np.zeros(FinalNy) + (NxConcatenate2Oct1stProfile*newDx)
    LidarProfile = np.zeros(FinalNx) + (195*newDy)
    
    fig, ax = plt.subplots(figsize=(10,6))
    plt.plot(xFinal,PeriodicProfile,':r',linewidth=2)
    plt.plot(OffshoreProfile,yFinal,'--m',linewidth=2)
    plt.plot(xFinal,LidarProfile,'-.w',linewidth=2)
    CS1 = plt.contour(xFinal, yFinal,FinalDepth,21, linewidths=0.5)
    plt.clabel(CS1,inline=1, fontsize=8) # inline controls labels
    
    CS2 = plt.pcolor(xFinal, yFinal,FinalDepth, cmap = 'viridis_r')
    plt.colorbar(CS2) # draw colorbar
    plt.legend(['Periodic Boundary Extension','Offshore Extension','LIDAR'])
    plt.axis([0,FinalLx,0,FinalLy])
    fig_name = os.path.join(ProjectDir,'2D_profiles','depth2D_Oct1stSurvey_%smglob_%snglob_%s.png'%(FinalNx,FinalNy,date))
    print('Saving 2D image:',fig_name)
    plt.savefig(fig_name, dpi=300)
    plt.close(fig)
    
    
    ## Save 2D Bathy
    fileOut = 'depth2D_Oct1stSurvey_%smglob_%snglob_%s.txt'%(FinalNx,FinalNy,date)
    funwaveFilePath = os.path.join(ProjectDir,'2D_profiles',fileOut)
    print('Saving 2D File:',funwaveFilePath,'\n')
    np.savetxt(funwaveFilePath, FinalDepth, fmt='%.6e', delimiter=' ', newline='\n', header='', footer='', comments='! ')
```
